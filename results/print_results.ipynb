```python
from matplotlib import pyplot as plt
import numpy as np


a = np.load('Transformer_ADAM_test.npy')
b = np.load('Transformer_SGDmom_test.npy')
c = np.load('Transformer_Vanilla_test.npy')
d = np.load('FFN_test.npy')
e = np.load('Transformer_small_test.npy')
f = np.load('Transformer_big_test.npy')

fig = plt.figure(figsize=(15, 15))

ax = fig.add_subplot(111)
plt.ion()

fig.show()
fig.canvas.draw()

ax.plot([np.mean(i) for i in c], label='Test Loss Vanila Transformer model')
ax.plot([np.mean(i) for i in b], label='Test_loss Transformer SGD with momentum')
ax.plot([np.mean(i) for i in a], label='Test_loss Transformer ADAM')
ax.plot([np.mean(i) for i in d], label='Test Loss Feed Forward Network')
ax.plot([np.mean(i) for i in e], label='Test Loss Transformer Small Model')
ax.plot([np.mean(i) for i in f], label='Test Loss Transformer Big Model')

ax.set_title("Comparing Mean Squared Error different Transformer Models")
ax.set_xlabel('Epoch')
ax.set_ylabel('Mean Squared Error')
ax.legend()
fig.canvas.draw()
```
