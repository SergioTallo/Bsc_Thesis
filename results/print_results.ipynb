```python
from matplotlib import pyplot as plt
import numpy as np


a = np.load('sequence 30/Transformer_SGDmom_test.npy')
b = np.load('Other sequences/Transformer_longseq_test.npy')
c = np.load('Other sequences/Transformer_veryshort_test.npy')
d = np.load('Other sequences/Transformer_shortseq_test.npy')
e = np.load('Other sequences/Transformer_verylong_test.npy')
f = np.load('Other sequences/Transformer_20_test.npy')
g = np.load('Other sequences/Transformer_10_test.npy')

fig = plt.figure(figsize=(15, 15))

ax = fig.add_subplot(111)
plt.ion()

fig.show()
fig.canvas.draw()

ax.plot([np.mean(i) for i in a], label='Test Loss sequence length 30')
ax.plot([np.mean(i) for i in e], label='Test Loss sequence length 120')
ax.plot([np.mean(i) for i in b], label='Test Loss sequence length 60')
ax.plot([np.mean(i) for i in f], label='Test Loss sequence length 20')
ax.plot([np.mean(i) for i in d], label='Test Loss sequence length 15')
ax.plot([np.mean(i) for i in g], label='Test Loss sequence length 10')
ax.plot([np.mean(i) for i in c], label='Test Loss sequence length 2')

ax.set_title("Comparing Mean Squared Error Vanilla Model with smaller model")
ax.set_xlabel('Epoch')
ax.set_ylabel('Mean Squared Error')
ax.legend()
fig.canvas.draw()
```
